# Calcul des Anomalies Mensuelles de la Couverture Nuageuse, de l'Amplitude Thermique Diurne, des Jours de Gel et de l'Évapotranspiration Potentielle par Site Minier (1901-2023) et par Rapport à la Période de Référence (1901-1950)

```python
import pandas as pd

# 📂 Charger les données climatiques
file_path = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extration of climate data\mine_climate_cld_dtr_frs_pet_1901_2023.csv"
df = pd.read_csv(file_path, parse_dates=["Date"])

# 🔄 Convertir le format de date en "YYYY-MM"
df["Date"] = df["Date"].dt.strftime("%Y-%m")

# 🗂 Extraire l'année et le mois
df["Annee"] = df["Date"].str[:4].astype(int)
df["Mois"] = df["Date"].str[5:7].astype(int)

# 🏛 Définir la période de référence (1901-1950)
periode_ref = df[(df["Annee"] >= 1901) & (df["Annee"] <= 1950)]

# 📊 Calcul de la moyenne historique mensuelle **pour chaque mine séparément**
moyennes_historiques = {}
variables = ["Cloud_Cover", "Diurnal_Temp_Range", "Frost_Days", "Potential_Evapotranspiration"]

for var in variables:
    moyennes_historiques[var] = (
        periode_ref.groupby(["Mine", "Mois"])[var]
        .mean()
        .reset_index()
        .rename(columns={var: f"{var}_historique"})
    )

    # 🔄 Fusionner chaque moyenne historique indépendamment dans le dataframe principal
    df = df.merge(moyennes_historiques[var], on=["Mine", "Mois"], how="left")

    # ➗ Calculer l’anomalie **spécifique à chaque mine**
    df[f"Anomalie_{var}"] = df[var] - df[f"{var}_historique"]

# 📂 Sauvegarde des résultats avec Date en "YYYY-MM"
output_path = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extration of climate data\anomalies_climatiques.csv"
df.to_csv(output_path, index=False)

# 📊 Afficher un aperçu des résultats
print(df[["Date", "Mine", "Mois"] + [f"{var}_historique" for var in variables] + [f"Anomalie_{var}" for var in variables]].head())
```

# Fusion des Données d'Anomalies Climatiques et de Production Minière avec Géolocalisation, Normalisation des Dates et Filtrage Temporel (2003-2023)

```python
import pandas as pd
import os

# 📂 Liste des fichiers d'anomalies climatiques
anomalies_files = [
    r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extration of climate data\anomalies_climatiques.csv",
    r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extration of climate data\anomalies_climatiques_pre_temp.csv",
    r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extration of climate data\anomalies_climatiques_tmn_tmx_vap_wet.csv"
]

# 📂 Fichier de production minière
production_file = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extraire coordonnées géo_mines\données_geocodés\données_geocodés\Données géocodées\données_geocodés\merged_data.csv"

# 📌 Initialisation du DataFrame des anomalies climatiques
df_anomalies = None

# 🔄 Parcourir les fichiers d'anomalies et les fusionner
for file in anomalies_files:
    if os.path.exists(file):
        try:
            df_temp = pd.read_csv(file, parse_dates=["Date"])
        except Exception:
            print(f"⚠️ Problème détecté avec {file}. Tentative avec ';' comme séparateur.")
            try:
                df_temp = pd.read_csv(file, sep=";", parse_dates=["Date"])
                print(f"✅ Fichier {file} corrigé !")
                df_temp.to_csv(file, index=False, sep=",", encoding="utf-8")
            except Exception as e:
                print(f"❌ Impossible de corriger {file} : {e}")
                continue

        # ✅ Vérifier si la colonne Date existe
        if "Date" not in df_temp.columns:
            print(f"❌ Problème : Colonne 'Date' absente dans {file}")
            continue

        # ✅ Convertir la colonne Date au format "YYYY-MM"
        df_temp["Date"] = df_temp["Date"].dt.strftime("%Y-%m")

        # 📌 Fusionner avec le DataFrame final
        if df_anomalies is None:
            df_anomalies = df_temp
        else:
            df_anomalies = df_anomalies.merge(df_temp, on=["Mine", "Latitude", "Longitude", "Date"], how="outer")

print("\n✅ Fusion des anomalies terminée !")
print(df_anomalies.head())

# 📂 Charger les données de production minière
if os.path.exists(production_file):
    df_production = pd.read_csv(production_file)

    # 📌 Vérifier si "Period" existe et le convertir en "Date" (YYYY-MM)
    if "Period" in df_production.columns:
        df_production.rename(columns={"Period": "Date"}, inplace=True)
        df_production["Date"] = pd.to_datetime(df_production["Date"], format="%b-%Y", errors="coerce")
        df_production["Date"] = df_production["Date"].dt.strftime("%Y-%m")  # ✅ Convertir en YYYY-MM

    # 📌 Fusion avec les anomalies climatiques
    df_final = df_production.merge(df_anomalies, on=["Mine", "Latitude", "Longitude", "Date"], how="outer")

    # 🔥 **Filtrer les mines sans coordonnées (Latitude ou Longitude manquante)**
    df_final = df_final.dropna(subset=["Latitude", "Longitude"])

    # 🔥 **Filtrer les données pour chaque mine : garder uniquement de 2003 à 2023**
    df_filtered = df_final[(df_final["Date"] >= "2003-01") & (df_final["Date"] <= "2023-12")]

    # 📊 **Calculer le nombre total de mines restantes**
    nombre_total_mines = df_filtered["Mine"].nunique()

    # 📂 Sauvegarde du fichier fusionné
    output_file = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\merged_anomalies_production_2003_2023.csv"
    df_filtered.to_csv(output_file, index=False, encoding="utf-8")
    
    # 📊 **Affichage des résultats**
    print(f"\n✅ Fichier fusionné sauvegardé : {output_file}")
    print(f"🔢 Nombre total de mines restantes dans la base : {nombre_total_mines}")

else:
    print(f"❌ Fichier de production minière introuvable : {production_file}")
```

# Fusion Finale des Données : Intégration des Prix et des Exportations avec la Production Minière, Anomalies Climatiques et Géolocalisation pour une Analyse  Complète (2003-2023)

```python
import pandas as pd

# 📂 Chemins des fichiers
file_merged_production = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\merged_anomalies_production_2003_2023.csv"
file_price = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\price_2003_2023.xlsx"
file_export = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\12_02_2025__21_48_41_exports.xls"

# 📌 1️⃣ Charger le fichier principal (Production et Anomalies Climatiques)
df_final = pd.read_csv(file_merged_production, parse_dates=["Date"], encoding="latin1")

# ✅ Convertir la colonne Date en format "YYYY-MM"
df_final["Date"] = df_final["Date"].dt.strftime("%Y-%m")

# 📌 2️⃣ Charger la base des prix
df_price = pd.read_excel(file_price, parse_dates=["Date"])

# ✅ Convertir la colonne Date en format "YYYY-MM"
df_price["Date"] = df_price["Date"].dt.strftime("%Y-%m")

# 🔹 **Correction uniquement pour Zaldívar**
df_final["Mine"] = df_final["Mine"].replace("ZaldÃ­var", "Zaldívar")
df_price["Mine"] = df_price["Mine"].replace("ZaldÃ­var", "Zaldívar")

# 📌 3️⃣ Charger la base des exportations
df_export = pd.read_excel(file_export)

# ✅ Renommer et convertir la colonne date
if "Period" in df_export.columns:
    df_export.rename(columns={"Period": "Date"}, inplace=True)
df_export["Date"] = pd.to_datetime(df_export["Date"], format="%b-%Y", errors="coerce").dt.strftime("%Y-%m")

# 📌 4️⃣ Fusionner les données
# 🔹 Fusion des prix (prise en compte de Mine, Latitude, Longitude, Date)
df_final = df_final.merge(df_price, on=["Mine", "Latitude", "Longitude", "Date"], how="left")

# 🔹 Fusion des exportations (uniquement sur Date)
df_final = df_final.merge(df_export, on="Date", how="left")

# 📌 5️⃣ Vérifier si **Zaldívar** a des prix après fusion
missing_prices_zaldivar = df_final[(df_final["Mine"] == "Zaldívar") & (df_final["prix_lme"].isna() | df_final["prix_comex"].isna())]

if not missing_prices_zaldivar.empty:
    print("⚠️ Problème : Zaldívar n'a pas de prix après fusion. Vérifiez la correspondance des dates !")
    print(missing_prices_zaldivar[["Mine", "Latitude", "Longitude", "Date"]].head())
else:
    print("✅ Zaldívar a bien des prix après fusion !")

# 📂 6️⃣ Sauvegarder le fichier fusionné
output_file = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\final_merged_data.csv"
df_final.to_csv(output_file, index=False, encoding="utf-8")

print(f"\n✅ Fusion terminée ! Fichier sauvegardé : {output_file}")
```
