# Découpage d'un Fichier KML Contenant une Grille de 5° × 5° en une Grille Finer de 0.1° × 0.1° (≈10 km × 10 km)

```python
import geopandas as gpd
from shapely.geometry import box

# Charger le fichier KML original
kml_file = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\extraire coordonnées géo_mines\cruts_4.08_gridboxes.kml"
gdf = gpd.read_file(kml_file, driver="LIBKML")

# Définir la taille des nouvelles cellules en degrés (~10 km x 10 km)
cell_size = 0.1  # ≈ 11 km x 11 km

# Créer une nouvelle liste pour stocker les petites cellules
new_cells = []

# Parcourir chaque grande cellule (5° x 5°)
for _, row in gdf.iterrows():
    minx, miny, maxx, maxy = row.geometry.bounds  # Coordonnées de la grande cellule
    
    # Découper en cellules plus petites
    x_range = list(range(int(minx * 10), int(maxx * 10), int(cell_size * 10)))
    y_range = list(range(int(miny * 10), int(maxy * 10), int(cell_size * 10)))
    
    for x in x_range:
        for y in y_range:
            small_cell = box(x / 10, y / 10, (x + cell_size * 10) / 10, (y + cell_size * 10) / 10)
            new_cells.append(small_cell)

# Créer un GeoDataFrame avec les nouvelles cellules
new_gdf = gpd.GeoDataFrame(geometry=new_cells, crs=gdf.crs)

# Sauvegarder en KML
output_kml = "reduced_10km_grid.kml"
new_gdf.to_file(output_kml, driver="KML")

print(f"✅ Grille réduite enregistrée dans {output_kml}")
```

# Chargement et Conversion des Grilles Réduites : De KML à Shapefile (SHP)

```python
import geopandas as gpd
import fiona

# ✅ Enable all Fiona drivers, including KML
fiona.supported_drivers['KML'] = 'rw'

# ✅ Load the KML file
kml_file = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\reduced_10km_grid.kml"
gdf = gpd.read_file(kml_file, driver="KML")

print("✅ Fichier KML chargé avec succès !")

# ✅ Convert to Shapefile (SHP)
output_shp = r"C:\Users\tchio\OneDrive\Bureau\data science\projet biscaye\reduced_10km_grid\reduced_10km_grid.shp"
gdf.to_file(output_shp, driver='ESRI Shapefile')

print("✅ Conversion réussie ! Le fichier Shapefile est sauvegardé sous :", output_shp)
```
